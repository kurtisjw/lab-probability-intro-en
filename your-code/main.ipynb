## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

##### YOUR CODE HERE

* Taking a white ball out.

There are 10 white balls in a bag of 30 balls, therefore there is a 10 in 30 (or 1 in 3) chance of taking a white ball out.

$\frac{10}{30}$

* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.


$\frac{10}{30}$ x $\frac{8}{29} $


* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.

$\frac{10}{30}$ = $\frac{1}{3}$

$\frac{8}{30}$ = $\frac{4}{15}$

$\frac{1}{3}$ x $\frac{4}{15}$ = $\frac{4}{45}$


* Taking 3 balls and them being all different colours.

For example, if we take a white ball, then a red ball, then a black ball:

$\frac{10}{30}$ x $\frac{12}{29}$ x $\frac{8}{28}$ = $\frac{960}{24360}$

There are six such sequences, so the probability is:

6 x $\frac{960}{24360}$ = 0.23


```python
# Challenge 1 with python

total_balls = 10 + 12 + 8

# Probability of taking a white ball out.
prob_white = 10/total_balls
print(prob_white)

#Probability of taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
prob_white = 10/total_balls
prob_black_after_white = 8/(total_balls - 1)
prob_white_then_black_without_replace = prob_white * prob_black_after_white
print(prob_white_then_black_without_replace)

# Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.

prob_white = 10/total_balls
prob_black_after_white = 8/total_balls
prob_white_then_black_with_replace = prob_white * prob_black_after_white
print(prob_white_then_black_with_replace)

# Taking 3 balls and them being all different colours.

prob_white = 10/total_balls
prob_red_after_white_no_replace = 12/(total_balls-1)
prob_black_after_white_red_no_replace = 8/(total_balls-2)
prob_W_R_B = prob_white * prob_red_after_white_no_replace * prob_black_after_white_red_no_replace

# There are six possible ways that 3 different could be withdrawn being all different colours, each with the same probability.
# Therefore, we multiply by 6

prob_all_diff = prob_W_R_B * 6
print(prob_all_diff)
```

#Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend.
To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy.
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
# YOUR CODE HERE

prob_cloudy_day = 0.4
prob_non_cloudy_day = 0.6
prob_rain_if_cloudy = 0.5
prob_rain_if_not_cloudy = 0.2

prob_rain = (prob_cloudy_day * prob_rain_if_cloudy) + (prob_non_cloudy_day * prob_rain_if_not_cloudy)

print(f"The probability of rain next saturday is: {prob_rain}")
```

## Challenge 3

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

```python
# YOUR CODE HERE

P_A = 0.3
P_B = 0.6
P_A_and_B = 0.1 #This represents both A and B occuring simultaneously

P_not_A = 1 - P_A
P_A_union_B = P_A + P_B - P_A_and_B
P_A_and_not_B = P_A - P_A_and_B
P_A_union_not_A = P_A + P_not_A
P_A_B_union_not_B = P_A = 0.3   # (Probability of A plus either B or not B, which covers all possibilities)
P_not_A_and_not_B = 1 - P_A_union_B  # Probability of A not occuring and B not occuring
P_not_A_and_B = 1 - P_A_and_B # Represents the probability that either A does not occur, B does not occur, or both do not occur


print(f"P(¬A) = {P_not_A}")
print(f"P(A ∪ B) = {P_A_union_B}")
print(f"P(A ∩ ¬B) = {P_A_and_not_B}")
print(f"P(A ∪ ¬A) = {P_A_union_not_A}")
print(f"P(A ∩ (B ∪ ¬B)) = {P_A_B_union_not_B}")
print(f"P(¬A ∩ ¬B) = {P_not_A_and_not_B}")
print(f"P(¬(A ∩ B)) = {P_not_A_and_B}")
```

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

```python
# YOUR CODE HERE

prob_covid_given_china = 0.01
prob_covid_given_italy = 0.2
prob_rolling_1 = 1/6 #Probability of rolling 1 and flying to Italy
prob_not_rolling_1 = 5/6 #Probability of not rolling 1 and flying to China

# Probability of getting COVID
probability_covid = (prob_covid_given_china * prob_not_rolling_1) + (prob_covid_given_italy * prob_rolling_1)

# Probabibility that I flew to China and got COVID
prob_china_given_covid = (prob_covid_given_china * prob_not_rolling_1) / probability_covid

print(f"The probability that I flew to China is {prob_china_given_covid}")
```
